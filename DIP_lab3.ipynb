```python
import cv2
import numpy as np
import math
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/drive/MyDrive/Colab Notebooks/Capture.PNG",0)
cv2_imshow(img)

c = (img/255)**5
transformedImg = 255*c

cv2_imshow(transformedImg)
```

```python
img=cv2.imread("/content/drive/MyDrive/Colab Notebooks/Capture1.PNG",0)
cv2_imshow(img)

# normalize float versions
norm_img = cv2.normalize(img, None, alpha=0, beta=1.4, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

norm_img = (255*norm_img).astype(np.uint8)

cv2_imshow(norm_img)
```

```python
img=cv2.imread("/content/drive/MyDrive/Colab Notebooks/Capture1.PNG",0)
cv2_imshow(img)

max = np.max(img[:])
min = np.min(img[:])
print(max,min)

img1=255*((img-min)/(max-min))
cv2_imshow(img1)
```

```python
img=cv2.imread("/content/drive/MyDrive/Colab Notebooks/3.png",0)
cv2_imshow(img)

max = np.max(img[:])
min = np.min(img[:])
print(max,min)

img1=255*((img-min)/(max-min))
cv2_imshow(img1)
```
